**Pencil Sketch with Python**

Creating the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

*Import Library*

```python
import cv2
import tensorflow as tf
```

*Read the image*

```python
from google.colab.patches import cv2_imshow
image = cv2.imread("/content/drive/MyDrive/powder blue room.jpg")
cv2_imshow( image)
```

*Converting the original image to greyscale*

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

*Invert the new grayscale image*

```python
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
```

*Process is to blur the image by using the Gaussian Function*

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
cv2_imshow(blurred)
```

*Invert the blurred image* 
Convert the image into a pencil sketch

```python
inverted_blurred = 255 - blurred
cv2_imshow(inverted_blurred)
```

*Converting it to a Pencil Sketch*

```python
pencil_sketch = cv2.divide(inverted_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
```

*Displaying both the images together*

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
```
